```python
def download_kaggle(data_dictionary, base_path):
    dest = base_path / data_dictionary["folder"]
    dest.mkdir(parents=True, exist_ok=True)
    try:
        cache_path = Path(kagglehub.dataset_download(data_dictionary["link"]))
        copied = []
        for src in cache_path.rglob("*"):
            if src.is_file():
                shutil.copy2(src, dest / src.name)
                copied.append(src.name)
        print(f"{data_dictionary['name']} done.")
        print(f"{len(copied)} file(s): {copied}")
        return dest
    except Exception as e:
        print(f"{e}")
        return None
```

```python
def preview_dataset(data_dictionary, base_dir, n_rows = 3):
    folder    = base_dir / data_dictionary["folder"]
    csv_files = list(folder.glob("*.csv")) if folder.exists() else []

    print(f"\n{'─'*70}")
    print(f"Dataset : {data_dictionary['name']}")

    dfs = {}
    for csv_file in csv_files:
        try:
            df = pd.read_csv(csv_file)
            dfs[csv_file.name] = df
            print(f"\nFile    : {csv_file.name}")
            print(f"Shape   : {df.shape[0]:,} rows x {df.shape[1]} cols")
            print(f"Columns : {list(df.columns)}")
            display(df.head(n_rows))
        except Exception as e:
            print(f"Error reading {csv_file.name}: {e}")

    return dfs
```
